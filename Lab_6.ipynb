```python
#Task_01

class Student:
    ID = 0
    def __init__(self,name,dep,age,CG):
        self.name = name
        self.dep = dep
        self.age = age
        self.CG = CG
        Student.ID += 1
        self.student_id = Student.ID

    def get_details(self):
        print('ID: ',self.student_id)
        print('Name: ',self.name)
        print('Department: ',self.dep)
        print('Age: ',self.age)
        print('CGPA: ',self.CG)

    @classmethod
    def from_String(cls,info):
        name,dep,age,CG = info.split('-')
        info = Student(name,dep,age,CG)
        return info

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print('Instance variables are variables that are used to store data that is exclusive to a single instance and Variables shared by all instances of a class are known as class variables. While instance variables, such as our name, email, and salary, might be unique for each instance, class variables should be the same for each instance.')
print('Class methods are linked to the class but not to the object of the class, whereas instance methods require the creation of an object of the same class before being called.')
```

```python
#Task_02

class Assassin:
    assassination_count = 0
    def __init__(self,assassin_name,succ_r):
        self.assassin_name = assassin_name
        self.succ_r = succ_r
        Assassin.assassination_count += 1
        self.assassinations = Assassin.assassination_count
    @classmethod
    def failurePercentage(cls,assassin_name,succ_r):
        failure_percentage = cls(assassin_name,100-succ_r)
        return failure_percentage
    @classmethod
    def failureRate(cls,assassin_name,succ_r):
        failure_rate = cls(assassin_name,100-succ_r)
        return failure_rate

    def printDetails(self):
        print('Name: ',self.assassin_name)
        print('Success Rate: ',str(self.succ_r)+'%')
        print('Total number of Assassin: ',self.assassinations)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task_05

class Employee:
    def __init__(self,name,workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,joining_year):
        workingPeriod = 2021 - joining_year
        info = cls(name,workingPeriod)
        return info
    @staticmethod
    def experienceCheck(workingPeriod,gender_of_worker):
        if gender_of_worker == 'male' and workingPeriod <3:
            return 'He is not experienced'
        elif gender_of_worker == 'female' and workingPeriod <3:
            return 'She is not experienced'
        elif gender_of_worker == 'male' and workingPeriod > 3:
            return 'He is experienced'
        else:
            return 'She is experienced'


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task_06

class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += self.count
    @staticmethod
    def advantage():
        print('Laptops are portable')
    @classmethod
    def resetCount(cls,laptopCount = 0):
        cls.laptopCount = 0

lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task_07

class Cat:
    Number_of_cats = 0
    def __init__(self,color_of_cat,action_of_cat):
        self.color_of_cat = color_of_cat
        self.action_of_cat = action_of_cat
        Cat.Number_of_cats += 1
    @classmethod
    def no_parameter(cls,color_of_cat = None, action_of_cat = None):
        if color_of_cat == None and action_of_cat == None:
            color_of_cat = 'white'
            action_of_cat = 'sitting'
            info_of_cat_1 = cls(color_of_cat,action_of_cat)
            return info_of_cat_1
    @classmethod
    def first_parameter(cls,color_of_cat,action_of_cat = None):
        if action_of_cat == None:
            action_of_cat = 'sitting'
            info_of_cat_2 = cls(color_of_cat,action_of_cat)
            return info_of_cat_2
    @classmethod
    def second_parameter(cls,action_of_cat,color_of_cat = None):
        if color_of_cat == None:
            color_of_cat = 'Grey'
            info_of_cat_3 = cls(color_of_cat,action_of_cat)
            return info_of_cat_3
    def changeColor(self,color_of_cat):
        self.color_of_cat = color_of_cat
    def printCat(self):
        print(self.color_of_cat,'cat is',self.action_of_cat)

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task_08

class Cylinder:
    radius = 5
    height = 18
    def __init__(self,radius,height):
        self.radius = radius
        self.height = height
        print('Default radius=',Cylinder.radius,'and height',Cylinder.height)
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print('Updated radius=',self.radius,'and height',self.height)
    @classmethod
    def swap(cls,radius,height):
        info_of_1st_cy = cls(height,radius)
        return info_of_1st_cy
        print(info_of_1st_cy)
    @classmethod
    def changeFormat(cls,info_of_cy):
        radius,height = info_of_cy.split('-')
        info_of_cy_2 = cls(float(radius),float(height))
        return info_of_cy_2
    @staticmethod
    def area(radius,height):
        import math
        area_of_cylinder = (2*math.pi*radius*radius)+(2*math.pi*radius*height)
        print('Area: ',area_of_cylinder)
    @staticmethod
    def volume(radius,height):
        import math
        volume_of_cylinder = math.pi*radius*radius*height
        print('Volume',volume_of_cylinder)


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task_09

class Student:
    total_students = 0
    brac_students = 0
    other_uni_students = 0
    def __init__(self,student_name,department,university = None):
        self.student_name = student_name
        self.department = department
        if university == None:
            self.university = 'Brac University'
        else:
            self.university = university
        Student.total_students += 1
        if university == None:
            Student.brac_students += 1
        else:
            Student.other_uni_students += 1
    @classmethod
    def createStudent(cls,student_name,department,university = None):
        Student_1_info = cls(student_name,department,university)
        return Student_1_info
    @classmethod
    def printDetails(cls):
        print('Total Student(s): ',Student.total_students)
        print('Brac University Student(s): ',Student.brac_students)
        print('Other Institution Student(s): ',Student.other_uni_students)
    def individualDetail(self):
        print('Name: ',self.student_name)
        print('Department: ',self.department)
        print('Institution: ',self.university)


Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```

```python
#Task_10

class SultansDine:
    outlets = 0
    amount_sold = 0
    list_of_info = []
    @classmethod
    def details(cls):
        print('Total Number of branches: ',cls.outlets)
        print('Total Sell: ',cls.amount_sold)
        for i in SultansDine.list_of_info:
            i.moreDetails()
    def __init__(self,branch_name):
        self.branch_name = branch_name
        SultansDine.outlets += 1
        SultansDine.list_of_info.append(self)
    def sellQuantity(self,sellQuantity):
        self.sellQuantity = sellQuantity
        if sellQuantity < 10:
            self.sale_in_branch = sellQuantity * 300
        elif sellQuantity < 20:
            self.sale_in_branch = sellQuantity * 350
        else:
            self.sale_in_branch = sellQuantity * 400
        SultansDine.amount_sold += self.sale_in_branch
    def branchInformation(self):
        print('Branch Name: ',self.branch_name)
        print('Branch Sell: {} Taka '.format(self.sale_in_branch))
    def moreDetails(self):
        print('Branch Name: {}, Branch Sell: {} Taka'.format(self.branch_name,self.sale_in_branch))
        self.percentage_of_sale = (self.sale_in_branch/SultansDine.amount_sold) * 100
        print("Branch consists total sell's: {x:.2f}%".format(x = self.percentage_of_sale))


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
